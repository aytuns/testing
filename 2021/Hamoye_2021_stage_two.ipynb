## Dropping the Date Axis

```python
df=df.drop(['date'],axis =1)
```

## Normalizing the data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
normalised_df.head()
```

```python
normalised_df.describe().transpose()
```

## Performing Linear Regression on T2 and T6

```python
X1 = normalised_df[['T2']]
y1 = normalised_df['T6']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X1,y1, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train, y_train)
```

```python
predicted_values= linear_model.predict(x_test)
```

```python
from sklearn import metrics
```

### R^2 Score

```python
r2_score = metrics.r2_score(y_test, predicted_values)
round(r2_score, 2)
```

## Using the full data set

### Separating data into Training set, and Testing set

```python
X = normalised_df.drop('Appliances',axis=1)
y = normalised_df['Appliances']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train, y_train)
```

```python
predicted_values= linear_model.predict(x_test)
```

### Mean Absolute Error

```python
mae = metrics.mean_absolute_error(y_test, predicted_values)
round(mae,2)
```

### Residual Sum of Squares

```python
rss = np.sum(np.square(y_test-predicted_values))
round(rss, 2)
```

### Root mean squared Error

```python
rmse = np.sqrt(metrics.mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

### Coefficent of Determination

```python
r2_score = metrics.r2_score(y_test, predicted_values)
round(r2_score,2)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg = Ridge(alpha=0.4)
```

```python
ridge_reg.fit(x_train, y_train)
```

```python
predicted_values2= ridge_reg.predict(x_test)
```

### Performance of the Ridge Regression

```python
rmse = np.sqrt(metrics.mean_squared_error(y_test, predicted_values2))
round(rmse, 3)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lasso_reg = Lasso(alpha=0.001)
```

```python
lasso_reg.fit(x_train, y_train)
```

```python
predicted_values3= lasso_reg.predict(x_test)
```

### Performance of the Lasso Regression

```python
rmse = np.sqrt(metrics.mean_squared_error(y_test, predicted_values3))
round(rmse, 3)
```

## Comparing The Effects of Regularisation on Feature weights

```python
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')

final_weights
```
